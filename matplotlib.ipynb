```python
# Inserts which are zoomed in

fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 10)
axes1.set_ylim(0, 10000)
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.set_title('power of 4')

# Small axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 20)
axes2.set_xlabel('a')
axes2.set_ylabel('b')
axes2.set_title('Zoomed in')
```

```python
# Inserts which are zoomed in

fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 10)
axes1.set_ylim(0, 10000)
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.set_title('power of 4')

# Small axes
axes2 = fig.add_axes([1, 1, 2, 2])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 20)
axes2.set_xlabel('a')
axes2.set_ylabel('b')
axes2.set_title('Zoomed in')
```

```python
fig = plt.figure(dpi = 100, figsize=(3, 2))

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(a, b)
axes.set_title("Power of 4")

fig.savefig('new_vis.png', bbox_inches = "tight")
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2) # default = 1, 1

axes[0][0].plot(x, y)
axes[0][0].set_title("hehe")

axes[1][0].plot(a, b)
axes[1][0].set_title("haha")
axes[1][0].set_xlabel("x for (1, 0)")

fig.suptitle("super title")
fig.set_figwidth(10)

plt.tight_layout()

fig.savefig("subplots.png", bbox_inches='tight')
```

```python
x = np.arange(0, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, label="y = x")
ax.plot(x, x**2, label="y = x^2")

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, label="y = x")
ax.plot(x, x**2, label="y = x^2")

ax.legend(loc='lower right')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, label="y = x")
ax.plot(x, x**2, label="y = x^2")

ax.legend(loc=(1.05, 0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, label="y = x")
ax.plot(x, x**2, label="y = x^2")

ax.legend(loc=(-0.25, 0.5))
```
