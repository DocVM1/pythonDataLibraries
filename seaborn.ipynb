# seaborn

```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
sns.scatterplot(x="salary", y="sales", data=df)
```

```python
plt.figure(dpi=120, figsize=(8, 4))
sns.scatterplot(x="salary", y="sales", data=df)
```

```python
plt.figure(dpi=120, figsize=(8, 4))
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education") # categorical column as hue
```

```python
plt.figure(dpi=120, figsize=(8, 4))
sns.scatterplot(x="salary", y="sales", data=df, hue="salary") # continuous column as hue
```

https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
plt.figure(dpi=120, figsize=(10, 4))
sns.scatterplot(x="salary", y="sales", data=df, 
                hue="level of education", palette="Dark2") # categorical column as hue
```

```python
plt.figure(dpi=120, figsize=(10, 4))
sns.scatterplot(x="salary", y="sales", data=df, size="salary")
```

```python
plt.figure(dpi=120, figsize=(10, 4))
sns.scatterplot(x="salary", y="sales", data=df, s=100)
```

```python
plt.figure(dpi=120, figsize=(10, 4))
sns.scatterplot(x="salary", y="sales", data=df, s=100, alpha=0.2)
```

```python
plt.figure(dpi=120, figsize=(10, 4))
sns.scatterplot(x="salary", y="sales", data=df, style="level of education")
```

```python
plt.figure(dpi=120, figsize=(10, 4))
sns.scatterplot(x="salary", y="sales", data=df, style="level of education", hue="level of education")
```

```python
plt.figure(dpi=120, figsize=(8, 4))
sns.scatterplot(x="salary", y="sales", data=df, hue="level of education")

plt.savefig('snsvis.png')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages.head()
```

```python
sns.rugplot(data=sample_ages, x='age')
```

```python
sns.displot(data=sample_ages, x='age', bins = 30)
```

```python
sns.displot(data=sample_ages, x='age', bins = 30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age') # we get -ve age also
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100])
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.1)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.5)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=10) # high bandwidth normalizes data
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.1, shade=True)
```
